```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("titanic")
```

```python
df
```

```python
df=df.dropna(subset=['deck'])
```

```python
df
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
df.describe()
```

```python
df['survived'].value_counts().plot(kind='bar')
plt.title('test plot',fontsize=8,color='y')
plt.xlabel('Analysis')
plt.ylabel('values')
plt.show()
```

```python
df[['survived','pclass','age','sibsp','parch','fare']].mean().plot(kind='bar')
plt.title('test plot',fontsize=8,color='red')
plt.xlabel('columns')
plt.ylabel('Mean')
plt.show()
```

```python
new_df=df.select_dtypes(include=['number'])
new_df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
new_df=scaler.fit_transform(new_df)
```

```python
new_df
```

```python
row_means=np.mean(new_df,axis=1)
x=np.arange(len(row_means))
```

```python
plt.subplot(1,2,1)
plt.bar(x,row_means,color='gray')
plt.title("Row-wise Mean",color='red')
plt.xlabel("Row index")
plt.ylabel("Mean value")
plt.show()
```

```python
row_std=np.std(new_df,axis=1)
plt.subplot(1,2,1)
plt.bar(x,row_std,color='gray')
plt.title("Row-wise Mean",color='red')
plt.xlabel("Row index")
plt.ylabel("Mean value")
plt.show()
```

```python
df_melted = df[['survived','pclass','age','sibsp','parch','fare']].melt(var_name="feature", value_name="values")
plt.figure(figsize=(12,8))
sns.boxplot(x="feature", y="values", data=df_melted)
plt.title("Box plot of Features")
plt.show()
```

```python
corr = df.corr(numeric_only=True)  
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Titanic Features")
plt.show()
```

```python
corr = df.corr(numeric_only=True)  
plt.figure(figsize=(8,6))
sns.clustermap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Titanic Features")
plt.show()
```
